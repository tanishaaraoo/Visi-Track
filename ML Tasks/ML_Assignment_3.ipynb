```python
# Load
match_df = pd.read_csv('matchpairsDevTrain.csv')
mismatch_df = pd.read_csv('mismatchpairsDevTrain.csv')

# Label
match_df['label'] = 1
mismatch_df['label'] = 0

# Align columns
match_df['name.1'] = match_df['name']

# Concatenate
train_pairs_df = pd.concat([match_df, mismatch_df], ignore_index=True)

# Make paths
def make_image_path(name, num):
    return f"lfw-deepfunneled/{name}/{name}_{int(num):04d}.jpg"

train_pairs_df['image1_path'] = train_pairs_df.apply(lambda row: make_image_path(row['name'], row['imagenum1']), axis=1)
train_pairs_df['image2_path'] = train_pairs_df.apply(lambda row: make_image_path(row['name.1'], row['imagenum2']), axis=1)

# Drop invalid paths
train_pairs_df = train_pairs_df.dropna(subset=['image1_path', 'image2_path'])

print(train_pairs_df.head())
```

```python
import os

missing_paths = []
for p in train_pairs_df['image1_path'].tolist() + train_pairs_df['image2_path'].tolist():
    if not os.path.exists(p):
        missing_paths.append(p)

print(f"Missing paths: {len(missing_paths)}")
if missing_paths:
    print(missing_paths[:10])
```

```python
from PIL import Image
from facenet_pytorch import MTCNN, InceptionResnetV1
import torch

# Load models
device = 'cuda' if torch.cuda.is_available() else 'cpu'
mtcnn = MTCNN(image_size=160, device=device)
resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)

# Embedding extraction
def get_embedding(image_path):
    try:
        img = Image.open(image_path)
        face = mtcnn(img)
        if face is None:
            return None
        face = face.unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = resnet(face)
        return embedding.squeeze().cpu().numpy()
    except Exception as e:
        print(f"Error for {image_path}: {e}")
        return None
```

```python
from tqdm import tqdm
import numpy as np

X_train = []
y_train = []

for idx, row in tqdm(train_pairs_df.iterrows(), total=len(train_pairs_df)):
    emb1 = get_embedding(row['image1_path'])
    emb2 = get_embedding(row['image2_path'])

    if emb1 is not None and emb2 is not None:
        dist = np.linalg.norm(emb1 - emb2)
        X_train.append([dist])
        y_train.append(row['label'])

print(f"✅ Finished. Got {len(X_train)} samples.")
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train, y_train)

print("✅ Model trained!")
```

```python
match_test_df = pd.read_csv('/content/matchpairsDevTest.csv')
mismatch_test_df = pd.read_csv('/content/mismatchpairsDevTest.csv')

match_test_df['label'] = 1
mismatch_test_df['label'] = 0
match_test_df['name.1'] = match_test_df['name']

test_pairs_df = pd.concat([match_test_df, mismatch_test_df], ignore_index=True)
```

```python
test_pairs_df['image1_path'] = test_pairs_df.apply(lambda row: make_image_path(row['name'], row['imagenum1']), axis=1)
test_pairs_df['image2_path'] = test_pairs_df.apply(lambda row: make_image_path(row['name.1'], row['imagenum2']), axis=1)
```

```python
import os

def file_exists(path):
    return os.path.exists(path)
```

```python
test_pairs_df = test_pairs_df[
    test_pairs_df['image1_path'].apply(file_exists) &
    test_pairs_df['image2_path'].apply(file_exists)
]
```

```python
X_test = []
y_test = []

for idx, row in tqdm(test_pairs_df.iterrows(), total=len(test_pairs_df)):
    emb1 = get_embedding(row['image1_path'])
    emb2 = get_embedding(row['image2_path'])

    if emb1 is not None and emb2 is not None:
        dist = np.linalg.norm(emb1 - emb2)
        X_test.append([dist])
        y_test.append(row['label'])

print(f"✅ Test set ready with {len(X_test)} samples.")
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

preds = clf.predict(X_test)
probs = clf.predict_proba(X_test)[:, 1]

print("✅ Test Accuracy:", accuracy_score(y_test, preds))
print("✅ Test ROC AUC:", roc_auc_score(y_test, probs))
```

```python
import numpy as np

def predict_match(img1_path, img2_path, model):
    # Get embeddings
    emb1 = get_embedding(img1_path)
    emb2 = get_embedding(img2_path)

    if emb1 is None or emb2 is None:
        print("Face not detected in one or both images.")
        return

    # Distance feature
    dist = np.linalg.norm(emb1 - emb2)
    print(f"Euclidean Distance: {dist:.4f}")

    # Predict
    prediction = model.predict([[dist]])[0]
    proba = model.predict_proba([[dist]])[0]

    if prediction == 1:
        print(f"Prediction: SAME PERSON (confidence {proba[1]:.2%})")
    else:
        print(f"Prediction: DIFFERENT PEOPLE (confidence {proba[0]:.2%})")
```
